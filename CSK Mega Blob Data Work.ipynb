```python
# Create df for eaach year on player-age df
player14_df = player_wanted_df.loc[player_wanted_df['season'] == '2014-15']
player15_df = player_wanted_df.loc[player_wanted_df['season'] == '2015-16']
player16_df = player_wanted_df.loc[player_wanted_df['season'] == '2016-17']
player17_df = player_wanted_df.loc[player_wanted_df['season'] == '2017-18']
player18_df = player_wanted_df.loc[player_wanted_df['season'] == '2018-19']

player15_df
```

```python
# Create df for eaach year on blob df
blob14_df = blob_df.loc[blob_df['season'] == 2014]
blob15_df = blob_df.loc[blob_df['season'] == 2015]
blob16_df = blob_df.loc[blob_df['season'] == 2016]
blob17_df = blob_df.loc[blob_df['season'] == 2017]
blob18_df = blob_df.loc[blob_df['season'] == 2018]

blob14_df
```

```python
# Merging each year 
only14_df = pd.merge(blob14_df, player14_df, on='player_name', how='inner')
only15_df = pd.merge(blob15_df, player15_df, on='player_name', how='inner')
only16_df = pd.merge(blob16_df, player16_df, on='player_name', how='inner')
only17_df = pd.merge(blob17_df, player17_df, on='player_name', how='inner')
only18_df = pd.merge(blob14_df, player18_df, on='player_name', how='inner')

only18_df
```

```python
# Append all dfs - FINAL df
df = only14_df.append(only15_df, ignore_index = True).append(only16_df, ignore_index = True).append(only17_df, ignore_index = True).append(only18_df, ignore_index = True)
```
