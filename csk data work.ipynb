```python
# Keeping specific columns and specific season only
nba_wanted_df = nba_df[['player_name', 'team_abbreviation', 'age','player_height', 'player_weight', 'season']]

nba_wanted_df = nba_wanted_df.loc[(nba_wanted_df['season'] == '2014-15') | (nba_wanted_df['season'] == '2015-16')
                                 | (nba_wanted_df['season'] == '2016-17') | (nba_wanted_df['season'] == '2017-18')
                                 | (nba_wanted_df['season'] == '2018-19') | (nba_wanted_df['season'] == '2019-20')]
```
